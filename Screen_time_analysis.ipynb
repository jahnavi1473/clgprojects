##Usage of apps analysis

```python
grouped_data = screen_data.groupby(['App', 'Date']).sum().reset_index()
```

```python
grouped_data
```

```python
fig,ax = plt.subplots(figsize=(40,20))
ax.bar(grouped_data['Date'], grouped_data['Usage'],width=0.5)
ax.set_xlabel('Date')
ax.set_ylabel('Usage')
ax.set_title('App usage by date')
plt.show()
```

```python
grouped_data['App'].value_counts()
```

```python
# Split data by app
app1_data =screen_data[screen_data['App'] == 'Instagram']
app2_data =screen_data[screen_data['App'] == 'Whatsapp']
```

```python
app1_data
```

```python
app2_data
```

```python
# Group data by date and sum usage
app1_usage_by_date = app1_data.groupby('Date')['Usage'].sum()
app2_usage_by_date = app2_data.groupby('Date')['Usage'].sum()
```

```python
app1_usage_by_date
```

```python
app2_usage_by_date
```

```python
merged = pd.merge(app1_usage_by_date, app2_usage_by_date, on='Date')
merged.columns = ['Instagram', 'Whatsapp']

# Plot bar graph
merged.plot(kind='bar', figsize=(10, 5))

# Add labels and title
plt.title('Usage Comparison Between instagram and whatsapp')
plt.xlabel('Date')
plt.ylabel('Usage')
plt.show()
```

Instagram usage is less than Whatsapp usage from the above visualization.

This implies whatsapp is the most used app

##Analysis based on number of notifications from the apps

```python
# Group data by date and sum notifications

app1_notifs_by_date = app1_data.groupby('Date')['Notifications'].sum()
app2_notifs_by_date = app2_data.groupby('Date')['Notifications'].sum()

merged = pd.merge(app1_notifs_by_date, app2_notifs_by_date, on='Date')
merged.columns = ['Instagram', 'Whatsapp']

# Plot bar graph
merged.plot(kind='bar', figsize=(10, 5))

# Add labels and title
plt.title('Notification Comparison Between instagram and whatsapp')
plt.xlabel('Date')
plt.ylabel('Notifications')
plt.show()
```

This visualization shows number of notifications for WhatsApp are more than Instagram notifications.

Analysis based on number of times app has been opened.

```python
# Group data by date and sum notifications

app1_timeop_by_date = app1_data.groupby('Date')['Times opened'].sum()
app2_timeop_by_date = app2_data.groupby('Date')['Times opened'].sum()

merged = pd.merge(app1_timeop_by_date, app2_timeop_by_date, on='Date')
merged.columns = ['Instagram', 'Whatsapp']

# Plot bar graph
merged.plot(kind='bar', figsize=(10, 5))

# Add labels and title
plt.title('Comparison Between instagram and whatsapp on times opened')
plt.xlabel('Date')
plt.ylabel('Times opened')
plt.show()
```

usage of smartphones is when we get notified by any app so the relationship between the number of notifications and the amount of usage is:

```python
fig,ax = plt.subplots(figsize=(20,10))
ax.scatter(grouped_data['Notifications'], grouped_data['Usage'],color='red')
m, b = np.polyfit(grouped_data['Notifications'], grouped_data['Usage'], 1)
plt.plot(grouped_data['Notifications'], m * grouped_data['Notifications'] + b)
ax.set_xlabel('Notifications')
ax.set_ylabel('Usage')
ax.set_title('Relationship between notifications and app usage')
plt.show()
```

There is a linear relationship between number of notifications and usage of apps. Hence, more the notifications more the usage of smartphones, more the screen time.
